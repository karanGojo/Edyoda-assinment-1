```python
# Python program to get Fibonacci series fronm 0 to 50
no = int(input("Enter the series of number _term:"))
n_1= int(input("enter the first no.:"))
n_2=int(input("enter the second no.:"))
count= 0
# condition to check the valid number
if no <=0:
    print("plz enter positive no.")
# if der is only one term den
elif no==1:
    print(n_1)
else:
    while count < no:
        print(n_1)
        fibo = n_1 + n_2
        n_1 = n_2# swapping method
        n_2 = fibo
        count+=1
        print(f"fibonacci for series:{fibo}")# formating
```
